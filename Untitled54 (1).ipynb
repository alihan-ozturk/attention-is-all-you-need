```python
import torch
from torch import nn
import numpy as np
import matplotlib.pyplot as plt
from torch import optim
from tqdm import tqdm
import sys
import math
```

```python
def sqlData(var, n, e):
    data = []
    str, stp = np.random.randint(0, 20, 2)
    x = np.linspace(str, stp, n)
    if var>2:
        data.append(np.linspace(str, stp, n))
        data.append(np.random.normal(0, np.random.rand(), n))
        var -= 2
    for i in range(var):
        stra, stpa = np.random.randint(-n//2, n//2, 2)
        x += np.sin(np.linspace(str+str, stp+stpa, n))*stra + i + 2
        data.append(x.copy())
    return data
```

```python
def dataset(n, batchsize, func, func_args):
    out = np.array([np.array(func(*func_args)) for _ in range(batchsize)])
    rnd = np.random.randint(0, out.shape[2]//4, n//batchsize)
    sql = np.random.randint(out.shape[2]//4, out.shape[2]*3//4, n//batchsize)
    for i in range(n//batchsize):
        src = out[: , :, rnd[i]:rnd[i]+sql[i]-1]
        tgt = out[:, :, rnd[i]+sql[i]] # .reshape(-1, 4, 1)
        yield src, tgt
```

```python
class PositionalEncoding(nn.Module):

    def __init__(self, var:int, dropout: float = 0.0, max_len: int = 5000):
        super().__init__()
        self.dropout = nn.Dropout(p=dropout)
        
        pe = torch.zeros(max_len, 1, var)
        for i in range(var):
            pe[:, 0, i] = torch.linspace(0, 1, max_len)
        self.register_buffer('pe', pe)

    def forward(self, x):
        x = x + self.pe[:x.size(0)]
        return self.dropout(x)


class transformerNNDuel(nn.Module):
    def __init__(self, var, nEncoder, nDecoder, d_model):
        super().__init__()
        encoderLayers = []
        nEncoderS = np.linspace(var, d_model, nEncoder+1, dtype=np.int32)
        for i in range(len(nEncoderS)-1):
            encoderLayers.append(nn.Linear(nEncoderS[i], nEncoderS[i+1]))
            encoderLayers.append(nn.ReLU(inplace=True))
            
        self.encoder = nn.Sequential(*nn.ModuleList(encoderLayers))

        self.pos_encoder = PositionalEncoding(var)
        enc = nn.TransformerEncoderLayer(d_model=d_model, nhead=8, dropout=0)
        self.transformerEncoder = nn.TransformerEncoder(enc, num_layers=6)    
    
        decoderLayers = []
        nDecoderS = np.linspace(d_model, var, nDecoder+1, dtype=np.int32)
        for i in range(len(nDecoderS)-1):
            decoderLayers.append(nn.Linear(nDecoderS[i], nDecoderS[i+1]))
            if i == len(nDecoderS)-2:
                break
            decoderLayers.append(nn.ReLU(inplace=True))
            
        self.adv = nn.Sequential(*nn.ModuleList(decoderLayers))

        decoderLayers = []
        nDecoderS = np.linspace(d_model, 1, nDecoder+1, dtype=np.int32)
        for i in range(len(nDecoderS)-1):
            decoderLayers.append(nn.Linear(nDecoderS[i], nDecoderS[i+1]))
            if i == len(nDecoderS)-2:
                break
            decoderLayers.append(nn.ReLU(inplace=True))
            
        self.state = nn.Sequential(*nn.ModuleList(decoderLayers))

        self.s = None
        self.a = None
        
    @staticmethod
    def mask(sz):
        mask = (torch.triu(torch.ones(sz, sz)) == 1).transpose(0, 1)
        mask = mask.float().masked_fill(mask == 0, float('-inf')).masked_fill(mask == 1, float(0.0))
        return mask
    
    def forward(self, x):
        x = self.pos_encoder(x)
        x = self.encoder(x)
        x = self.transformerEncoder(x, mask=self.mask(len(x)))
        self.s = self.state(x) # 1
        self.a = self.adv(x) # seq
        x = self.s + (self.a - self.a.mean(2).unsqueeze(2))
        return x


class transformerNN(nn.Module):
    def __init__(self, var, nEncoder, nDecoder, d_model):
        super().__init__()
        encoderLayers = []
        nEncoderS = np.linspace(var, d_model, nEncoder+1, dtype=np.int32)
        for i in range(len(nEncoderS)-1):
            encoderLayers.append(nn.Linear(nEncoderS[i], nEncoderS[i+1]))
            encoderLayers.append(nn.ReLU(inplace=True))
            
        self.encoder = nn.Sequential(*nn.ModuleList(encoderLayers))

        self.pos_encoder = PositionalEncoding(var)
        enc = nn.TransformerEncoderLayer(d_model=d_model, nhead=8, dropout=0)
        self.transformerEncoder = nn.TransformerEncoder(enc, num_layers=6)    
    
        decoderLayers = []
        nDecoderS = np.linspace(d_model, var, nDecoder+1, dtype=np.int32)
        for i in range(len(nDecoderS)-1):
            decoderLayers.append(nn.Linear(nDecoderS[i], nDecoderS[i+1]))
            if i == len(nDecoderS)-2:
                break
            decoderLayers.append(nn.ReLU(inplace=True))
            
        self.decoder = nn.Sequential(*nn.ModuleList(decoderLayers))

        
    @staticmethod
    def mask(sz):
        mask = (torch.triu(torch.ones(sz, sz)) == 1).transpose(0, 1)
        mask = mask.float().masked_fill(mask == 0, float('-inf')).masked_fill(mask == 1, float(0.0))
        return mask
    
    def forward(self, x):
        x = self.pos_encoder(x)
        x = self.encoder(x)
        x = self.transformerEncoder(x, mask=self.mask(len(x)))
        x = self.decoder(x)
        return x


class vanillaNN(nn.Module):
    def __init__(self, var, nEncoder, nDecoder, d_model):
        super().__init__()
        encoderLayers = []
        nEncoderS = np.linspace(var, d_model, nEncoder+1, dtype=np.int32)
        for i in range(len(nEncoder)-1):
            encoderLayers.append(nn.Linear(nEncoderS[i], nEncoderS[i+1]))
            encoderLayers.append(nn.ReLU(inplace=True))
            
        self.encoder = nn.Sequential(*nn.ModuleList(encoderLayers))
    
        decoderLayers = []
        nEncoderS = np.linspace(d_model, var, nDecoder+1, dtype=np.int32)
        for i in range(len(nDecoder)-1):
            decoderLayers.append(nn.Linear(nEncoderS[i], nEncoderS[i+1]))
            if i == len(nEncoderS)-2:
                break
            decoderLayers.append(nn.ReLU(inplace=True))
            
        self.decoder = nn.Sequential(*nn.ModuleList(decoderLayers))

    
    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x
```

```python
epoch = 1000
var = 6
seqlen = 40
dmodel = 512
lr = 5e-5
enc = 6
dec = 6
batchSize = 96


net = transformerNNDuel(var, enc, dec, dmodel).to("cuda")
criterion = nn.MSELoss()
optimizer = optim.Adam(net.parameters(), lr=lr)
hist = []

minLoss = float('inf')
for i in range(epoch):
    net.train()
    for j, (src, tgt) in enumerate(dataset(batchSize*8, batchSize, sqlData, (var, seqlen, 0.2))):
        src = torch.from_numpy(src).to(torch.float32).permute(2, 0, 1).to("cuda")
        tgt = torch.from_numpy(tgt).to(torch.float32).to("cuda")
        
        optimizer.zero_grad()

        outputs = net(src)
        loss = criterion(outputs.sum(0), tgt)
        loss.backward()
        optimizer.step()
        
    net.eval()
    runningLoss = 0
    for k, (src, tgt) in enumerate(dataset(batchSize*2, batchSize, sqlData, (var, seqlen, 0.2))):
        src = torch.from_numpy(src).to(torch.float32).permute(2, 0, 1).to("cuda")
        tgt = torch.from_numpy(tgt).to(torch.float32).to("cuda")
        
        with torch.no_grad():
             outputs = net(src)

        loss = criterion(outputs.sum(0), tgt)

        runningLoss += loss.item()

    runningLoss = runningLoss/(k+1)
    
    if minLoss>runningLoss:
        minLoss = runningLoss
        bestWeights = net.state_dict().copy()
        
    hist.append(runningLoss)
    
    print(f'Current loss: {hist[-1]} Epoch: {i+1} ', end="\r")

net._load_param = bestWeights
```
