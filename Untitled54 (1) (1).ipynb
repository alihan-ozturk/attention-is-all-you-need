# TransformerALI: Adaptive Learning wint Integration Transformer

```python
def sqlData(var, n, e):
    data = []
    str, stp = np.random.randint(0, 20, 2)
    x = np.linspace(str, stp, n) + np.random.normal(0, 0.2, n)
    if var>2:
        data.append(x)
        data.append(np.random.normal(0, np.random.rand(), n))
        var -= 2
    x = np.zeros_like(x)
    for i in range(var):
        stra, stpa = np.random.randint(-n//2, n//2, 2)
        x += np.sin(np.linspace(stra-stpa//3, stra+stpa//3, n))*stpa + np.random.normal(0, 0.5, n)
        data.append(x.copy())
    return data
```

```python
def dataset(n, batchsize, func, func_args):
    out = np.array([np.array(func(*func_args)) for _ in range(batchsize)])
    rnd = np.random.randint(0, out.shape[2]//4, n//batchsize)
    sql = np.random.randint(out.shape[2]//4, out.shape[2]*3//4, n//batchsize)
    for i in range(n//batchsize):
        src = out[: , :, rnd[i]:rnd[i]+sql[i]]
        tgt = out[:, :, rnd[i]+1 + sql[i]] #  # .reshape(-1, 4, 1)
        yield src, tgt
```

```python
class PositionalEncoding(nn.Module):

    def __init__(self, var:int, dropout: float = 0.0, max_len: int = 1000):
        super().__init__()
        self.dropout = nn.Dropout(p=dropout)
        
        pe = torch.zeros(max_len, 1, var)
        for i in range(var):
            pe[:, 0, i] = torch.linspace(-1.77, 1.77, max_len) 
        self.register_buffer('pe', pe)
        self.max_len = max_len
        
    def forward(self, x):
        s = (self.max_len - x.size(0))/2
        x = x + self.pe[math.floor(s): -math.ceil(s)]
        return self.dropout(x)

def batchStandardize(x, params=None):
    if params == None:
        mean = x.mean(0)
        std = x.std(0)
        return (x - mean) / std, [mean, std] 
    else:
        return (x - params[0]) / params[1]


class transformerALI(nn.Module):
    def __init__(self, var, nEncoder, nDecoder, d_model):
        super().__init__()

        self.pos = PositionalEncoding(d_model, max_len=1000)

        self.encoder = self.sequentialLayer(var, d_model, nEncoder, nn.ReLU)
        
        enc = nn.TransformerEncoderLayer(d_model=d_model, nhead=16)
        self.transformerEncoder = nn.TransformerEncoder(enc, num_layers=16)    

        self.decoder = self.sequentialLayer(d_model, var, nDecoder, nn.ReLU)
        
    @staticmethod
    def sequentialLayer(inputFeatures, outputFeatures, nLayer, activationFunc):
        Layers = []
        nLay = np.linspace(inputFeatures, outputFeatures, nLayer+1, dtype=np.int32)
        for i in range(len(nLay)-1):
            Layers.append(nn.Linear(nLay[i], nLay[i+1]))
            if i == len(nLay)-2:
                break
            Layers.append(activationFunc(inplace=True))
        return nn.Sequential(*nn.ModuleList(Layers))
    
    @staticmethod
    def mask(sz):
        mask = (torch.triu(torch.ones(sz, sz)) == 1).transpose(0, 1)
        mask = mask.float().masked_fill(mask == 0, float('-inf')).masked_fill(mask == 1, float(0.0))
        return mask
    
    def forward(self, x):
        x = self.encoder(x)
        x = self.pos(x)
        x = self.transformerEncoder(x, self.mask(len(x)))
        x = self.decoder(x)
        return x.sum(0)

def updateWeightSeq(model, seq):
    for param in model.parameters():
        if param.requires_grad:
            param.grad /= seq
```

```python
epoch = 50000
var = 6
seqlen = 40
dmodel = 512
lr = 5e-8
batchSize = 96
nEncoder = 6
nDecoder = 6
```

```python
net = transformerALI(var, nEncoder, nDecoder, dmodel).to("cuda")
criterion = nn.MSELoss()
optimizer = optim.Adam(net.parameters(), lr=lr, weight_decay=0.9)
```

```python
hist = []

minLoss = float('inf')
for i in range(epoch):
    net.train()
    
    for j, (src, tgt) in enumerate(dataset(batchSize*20, batchSize, sqlData, (var, seqlen, 0.2))):
        src, srcParams = batchStandardize(torch.from_numpy(src).to(torch.float32).permute(2, 0, 1).to("cuda"))
        tgt = batchStandardize(torch.from_numpy(tgt).to(torch.float32).to("cuda"), srcParams)
        
        optimizer.zero_grad()

        outputs = net(src)
        loss = criterion(outputs, tgt) # *0.9 + criterion2(src, outputs)*0.1
        loss.backward()
        # updateWeightSeq(net, len(src)) # olmayınca daha iyi
        optimizer.step()
    
    net.eval()
    runningLoss = 0
    for k, (src, tgt) in enumerate(dataset(batchSize*4, batchSize, sqlData, (var, seqlen, 0.2))):
        src, srcParams = batchStandardize(torch.from_numpy(src).to(torch.float32).permute(2, 0, 1).to("cuda"))
        tgt = batchStandardize(torch.from_numpy(tgt).to(torch.float32).to("cuda"), srcParams)
        
        with torch.no_grad():
             outputs = net(src)

        loss = criterion(outputs, tgt)

        runningLoss += loss.item()
        
    runningLoss = runningLoss/(k+1)
    
    if minLoss>runningLoss:
        minLoss = runningLoss
        bestWeights = net.state_dict().copy()
        
    hist.append(runningLoss)
    
    print(f'Current loss: {hist[-1]} Epoch: {i+1} ', end="\r")

# net._load_param = bestWeights.copy()
```
