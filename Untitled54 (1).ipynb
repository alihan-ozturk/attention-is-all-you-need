```python
def sqlData(var, n, e):
    data = []
    str, stp = np.random.randint(0, 20, 2)
    x = np.linspace(str, stp, n)
    if var>2:
        data.append(x)
        data.append(np.random.normal(0, np.random.rand(), n))
        var -= 2
    x = np.zeros_like(x)
    for i in range(var):
        stra, stpa = np.random.randint(-n//2, n//2, 2)
        x += np.sin(np.linspace(str+str, stp+stpa, n))*stra * (i + 1) + 2
        data.append(x.copy())
    return data
```

```python
def dataset(n, batchsize, func, func_args):
    out = np.array([np.array(func(*func_args)) for _ in range(batchsize)])
    rnd = np.random.randint(0, out.shape[2]//4, n//batchsize)
    sql = np.random.randint(out.shape[2]//4, out.shape[2]*3//4, n//batchsize)
    for i in range(n//batchsize):
        src = out[: , :, rnd[i]:rnd[i]+sql[i]]
        tgt = out[:, :, rnd[i]+sql[i]+1] # .reshape(-1, 4, 1)
        yield src, tgt
```

```python
class PositionalEncoding(nn.Module):

    def __init__(self, var:int, dropout: float = 0.0, max_len: int = 5000):
        super().__init__()
        self.dropout = nn.Dropout(p=dropout)
        
        pe = torch.zeros(max_len, 1, var)
        for i in range(var):
            pe[:, 0, i] = torch.linspace(0, 1, max_len)
        self.register_buffer('pe', pe)

    def forward(self, x):
        x = x + self.pe[:x.size(0)]
        return self.dropout(x)

class transformerALI(nn.Module):
    def __init__(self, var, nEncoder, nDecoder, d_model):
        super().__init__()

        self.pos = PositionalEncoding(d_model, max_len=1000)
    
        self.encoder = self.sequentialLayer(var, d_model, nEncoder, nn.ReLU)
        
        enc = nn.TransformerEncoderLayer(d_model=d_model, nhead=8, dropout=0)
        self.transformerEncoder = nn.TransformerEncoder(enc, num_layers=6)    
    
        self.adv = self.sequentialLayer(d_model, var, nDecoder, nn.ReLU)
        self.state = self.sequentialLayer(d_model, var, nDecoder, nn.ReLU)

    @staticmethod
    def sequentialLayer(inputFeatures, outputFeatures, nLayer, activationFunc):
        nLay = np.linspace(inputFeatures, outputFeatures, nLayer+1, dtype=np.int32)
        nLayers = []
        for i in range(len(nLay)-1):
            nLayers.append(nn.Linear(nLay[i], nLay[i+1]))
            if i == len(nLay)-2:
                break
            nLayers.append(activationFunc(inplace=True))
        return nn.Sequential(*nn.ModuleList(nLayers))
        
    @staticmethod
    def mask(sz):
        mask = (torch.triu(torch.ones(sz, sz)) == 1).transpose(0, 1)
        mask = mask.float().masked_fill(mask == 0, float('-inf')).masked_fill(mask == 1, float(0.0))
        return mask
    
    def forward(self, x):
        x = self.encoder(x)
        x = self.pos(x)
        x = self.transformerEncoder(x, self.mask(len(x)))
        state = self.state(x.sum(0))
        adv = self.adv(x) # best score before decode doing average
        return state + adv.sum(0)


def updateWeightSeq(model, seq):
    for param in model.parameters():
        if param.requires_grad:
            param.grad *= seq
```

```python
epoch = 50000
var = 5
seqlen = 40
dmodel = 512
lr = 5e-6
batchSize = 96
nEncoder = 4
nDecoder = 4

# net = transformerFantasy(seq, var, enc, dec, dmodel).to("cuda")
net = transformerALI(var, nEncoder, nDecoder, dmodel).to("cuda")
criterion = nn.MSELoss()
# criterion2 = CorrelationLoss()
optimizer = optim.Adam(net.parameters(), lr=lr, weight_decay=0.9995)
hist = []

minLoss = float('inf')
for i in range(epoch):
    net.train()
    for j, (src, tgt) in enumerate(dataset(batchSize*8, batchSize, sqlData, (var, seqlen, 0.2))):
        src = torch.from_numpy(src).to(torch.float32).permute(2, 0, 1).to("cuda")
        tgt = torch.from_numpy(tgt).to(torch.float32).to("cuda")
        
        optimizer.zero_grad()

        outputs = net(src)
        loss = criterion(outputs, tgt) # *0.9 + criterion2(src, outputs)*0.1
        loss.backward()
        # updateWeightSeq(net, len(src)) olmayÄ±nca daha iyi
        optimizer.step()
        
    net.eval()
    runningLoss = 0
    for k, (src, tgt) in enumerate(dataset(batchSize*2, batchSize, sqlData, (var, seqlen, 0.2))):
        src = torch.from_numpy(src).to(torch.float32).permute(2, 0, 1).to("cuda")
        tgt = torch.from_numpy(tgt).to(torch.float32).to("cuda")
        
        with torch.no_grad():
             outputs = net(src)

        loss = criterion(outputs, tgt)

        runningLoss += loss.item()

    runningLoss = runningLoss/(k+1)
    
    if minLoss>runningLoss:
        minLoss = runningLoss
        bestWeights = net.state_dict().copy()
        
    hist.append(runningLoss)
    
    print(f'Current loss: {hist[-1]} Epoch: {i+1} ', end="\r")

# net._load_param = bestWeights.copy()
```
